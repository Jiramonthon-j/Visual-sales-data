```python
import matplotlib.pyplot as plt
```

```python

summary = df.groupby(['Year', 'Product Name'])['Total Sales'].sum().unstack()


summary.plot(kind='bar', figsize=(10, 6))


plt.title('ยอดขายสินค้าแต่ละชนิดในแต่ละปี', fontsize=16)
plt.xlabel('ปี (Year)', fontsize=12)
plt.ylabel('ยอดขายรวม (บาท)', fontsize=12)
plt.legend(title='ชื่อสินค้า (Product Name)')
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


ax = summary.plot(kind='bar', figsize=(10,6))

plt.title('ยอดขายสินค้าแต่ละชนิดในแต่ละปี', fontsize=16)
plt.xlabel('ปี (Year)', fontsize=12)
plt.ylabel('ยอดขายรวม (บาท)', fontsize=12)
plt.legend(title='ชื่อสินค้า (Product Name)')
plt.grid(axis='y', linestyle='--', alpha=0.7)


for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge', fontsize=9, padding=3)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


years = ['2023', '2024', '2025']
headphone = [120000, 157500, 198000]
laptop = [1575000, 1092000, 1350000]
mouse = [81000, 114000, 130000]

x = np.arange(len(years))
width = 0.25

fig, ax = plt.subplots(figsize=(10,6))
rects1 = ax.bar(x - width, headphone, width, label='Headphone H5')
rects2 = ax.bar(x, laptop, width, label='Laptop X1')
rects3 = ax.bar(x + width, mouse, width, label='Mouse M7')


for rects in [rects1, rects2, rects3]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')


ax.set_xlabel('Year', fontsize=12)          
ax.set_ylabel('Sales (baht)', fontsize=12)  
ax.set_title('Compare product sales each year', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.legend(title='Product')

plt.show()
```
